```python
import pandas as pd
```

```python
xls = pd.ExcelFile('T01.xlsx')
```

```python
all1 = pd.DataFrame([])
```

```python
for sh in xls.sheet_names:
    df = pd.read_excel('T01.xlsx', sheet_name = sh)
    df["month"] = sh
    all1 = all1.append(df, ignore_index=True)
print(all1)
```
